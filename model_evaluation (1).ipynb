### Exercise 4

#### (a)
Compute the average test error and the standard errors for each model. 

```python
import numpy as np

# Test errors for Model A and B
moda = np.array([0.13, 0.07, 0.08, 0.04, 0.11, 0.16, 0.10, 0.08, 0.10, 0.12])
modb = np.array([0.14, 0.09, 0.09, 0.045, 0.12, 0.16, 0.12, 0.09, 0.11, 0.125])
```

```python
# Print test errors
te_moda = np.mean(moda)
print('avg test error for Model A=',te_moda)
te_modb = np.mean(modb)
print('avg test error for Model B=',np.round_(te_modb,decimals = 3))
```

```python
# Print standard errors of test errors
se_moda = np.sqrt(np.var(moda,ddof=1)/len(moda))
print('SE of the avg test error for Model A=',np.round_(se_moda,decimals = 3))
se_modb = np.sqrt(np.var(modb,ddof=1)/len(modb))
print('SE of the avg test error for Model B=',np.round_(se_modb,decimals = 3))
```

*Note:* By setting the degrees of of freedom `ddof=1`, the sample variance is used to estimate the variance of the test errors.

```python
# Print 95% CI
b_moda = te_moda + np.array([- 2*se_moda,2*se_moda])
print('95% CI for the test error for Model A=',np.round_(b_moda,decimals = 3))
b_modb = te_modb + np.array([- 2*se_modb,2*se_modb])
print('95% CI for the test error for Model B=',np.round_(b_modb,decimals = 3))
```

#### (b)
Compute the average of the difference of the test errors between models A and B and the standard error. 

```python
# Print the average difference of the test errors
te_diff = np.mean(modb-moda)
print('avg difference of the test errors =',np.round_(te_diff,decimals = 3))
```

```python
# Print standard errors of the difference of the test errors
se_diff = np.sqrt(np.var(modb-moda,ddof=1)/len(moda))
print('SE of the difference of the test errors=',np.round_(se_diff,decimals = 3))
```

```python
# Print 95% CI
b_diff = te_diff + np.array([- 2*se_diff,2*se_diff])
print('95% CI for the difference of the test errors=',np.round_(b_diff,decimals = 3))
```

#### (d)
Create a plot that displays for $n=1,2,\ldots,1000$, the probability that the $n^{\text{th}}$ data point from the original  dataset is in the bootstrap sample

```python
import matplotlib.pyplot as plt

N = np.arange(1, 1001, 1)
fN = 1-(1-1/N)**N
plt.plot(N, fN, '-k')
```
